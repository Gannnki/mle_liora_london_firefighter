```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

#transformation Eastern Northing into Latitude, Longitude
from pyproj import Transformer

Inc = pd.read_csv("../data/processed/Inc.csv")
```

```python
transformer = Transformer.from_crs("EPSG:27700", "EPSG:4326", always_xy=True)

for i in Inc.index:
    if pd.isna(Inc.loc[i, 'Latitude']) or pd.isna(Inc.loc[i, 'Longitude']):
        easting = Inc.loc[i, 'Easting_rounded']
        northing = Inc.loc[i, 'Northing_rounded']
        
        if pd.notna(easting) and pd.notna(northing):
            lon, lat = transformer.transform(easting, northing)
            Inc.loc[i, 'Latitude'] = lat
            Inc.loc[i, 'Longitude'] = lon
```
