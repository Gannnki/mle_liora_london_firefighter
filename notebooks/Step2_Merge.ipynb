```python
# after isnpection a selection of the columns is made
#mob_cols = ["IncidentNumber", "CalYear", "HourOfCall", "DateAndTimeMobilised", "DeployedFromStation_Name", "Address", "Latitude", "Longitude"]
# We presume for the same incidente the data comes from the same font, so they have to be the same therefor

#Prepare the cloumns we want to keep for the merge
mob_cols = [
    "IncidentNumber",
    "CalYear",
    "HourOfCall",
    "DateAndTimeMobilised",
    "AttendanceTimeSeconds",
    "DeployedFromStation_Name",
    "PumpOrder",
    "Latitude",
    "Longitude"]

inc_cols = [
    "IncidentNumber",
    "DateOfCall",
    "IncidentGroup",
    "SpecialServiceType",
    "PropertyCategory",
    "PropertyType",
    "IncGeo_BoroughName",
    "Latitude",
    "Longitude",
    "NumCalls",
]

#Made a new Df with only the columns
Mob_sel = Mobac102[mob_cols].copy()
Inc_sel = Inc[inc_cols].copy()

#I still need to change the name of some columns
Mob_sel = Mob_sel.rename(columns={
    "Latitude": "Station_Latitude",
    "Longitude": "Station_Longitude"})

Inc_sel = Inc_sel.rename(columns={
    "Latitude": "Incident_Latitude",
    "Longitude": "Incident_Longitude"})



```

```python
Mob_sel.columns.unique()
```

```python
Inc_sel.columns.unique()
```

```python
#I prepare the Mob dfbecause of the diferents of numbers of Incident number in Inc and Mob

Mob_agg = Mob_sel.groupby("IncidentNumber").agg(
    Mob_CalYear=("CalYear", "first"),
    Mob_HourOfCall=("HourOfCall", "first"),
    DateAndTimeMobilised=("DateAndTimeMobilised", "first"),
    AttendanceTimeSeconds=("AttendanceTimeSeconds", "mean"),
    DeployedFromStation_Name=("DeployedFromStation_Name", lambda x: x.mode().iloc[0] if not x.mode().empty else None),
    PumpOrder=("PumpOrder", "max"),
    Station_Latitude=("Station_Latitude", "first"),
    Station_Longitude=("Station_Longitude", "first")
).reset_index()

#Merge of the 2 lists
Mob_Inc = Inc_sel.merge(Mob_agg, on="IncidentNumber", how="inner")

#Verification
print(Mob_Inc.shape)
print(Mob_Inc.columns.tolist())
Mob_Inc.head()
```
